Alright, unsupervised learning huh? Well what exactly is this? Basically when we don't exaclty know the labels to which we will devide our data on. 

For eg: A cloth manufacturer wants to decide on the sizes of the shirts he wanna manufacture on the basis of the height and weight of the people. 

Now the size can be:
- small, medium and large 
or it can be 
- extra small, small, medium, large, extra large. 

So by this you probably get an idea of what unsupervised learning is. 

We will understand more in depth once we dive into the algorithms with some data sets.


```python
# Import pandas and a dataset suitable for k-means clustering (Iris dataset without labels)
import pandas as pd

# Load the Iris dataset (features only, no labels for unsupervised learning)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
X = df.drop('species', axis=1)  # Remove the label column

# Display the first 5
X.head()
```

```python
# Perform k-means clustering on the Iris dataset (without labels)
from sklearn.cluster import KMeans

# Choose the number of clusters (k=3 for Iris)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

# Add the cluster labels to the dataframe
X['cluster'] = kmeans.labels_

# Display the first 5 rows with cluster labels
X.head()
```

```python
# Visualize the clusters formed by k-means on the Iris dataset
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(X['sepal_length'], X['sepal_width'], c=X['cluster'], cmap='viridis', s=50)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clusters on Iris Dataset')
plt.colorbar(label='Cluster')
plt.show()
```

Explanation of the K-Means Cluster Graph

The scatter plot visualizes the clusters formed by the k-means algorithm on the Iris dataset using the features **sepal length** and **sepal width**. Each point represents a flower sample, and the color indicates the cluster assigned by k-means.

- **Axes:**  
  - The x-axis shows the sepal length of each sample.
  - The y-axis shows the sepal width.

- **Colors:**  
  - Each color corresponds to a different cluster (as determined by k-means).  
  - Since k-means is unsupervised, these clusters do not directly correspond to the actual species, but often align well with the natural groupings in the data.

- **Interpretation:**  
  - You can observe how the algorithm has grouped the data into three clusters, which often correspond to the three Iris species.
  - Overlapping regions may indicate samples that are similar in sepal length and width, making them harder to separate.
